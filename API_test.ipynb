```python
countries = ['ARG', 'CHL', 'BOL', 'VEN']
latam_df = df[(df['source_iso3'].isin(countries)) & (df['target_iso3'].isin(countries))]
```

## Analisis mobilidad y migraciÃ³n hacia Argentina

```python
# Filtrar los datos para obtener solo los viajes desde CHL, BOL y VEN hacia ARG
filtered_data_arg = latam_df[(latam_df['target_iso3'] == 'ARG') & (latam_df['source_iso3'].isin(['CHL', 'BOL', 'VEN']))]

filtered_data_arg = filtered_data_arg[['estimated_trips', 'year', 'source_iso3']]

filtered_data_arg
```

```python
pivot_AR = pd.pivot_table(filtered_data_arg, values='estimated_trips', index='source_iso3', columns='year', aggfunc='sum')

pivot_AR
```
