```python
import pandas as pd
import requests
import json

# Declares the base url for calling the API
base_url = "https://population.un.org/dataportalapi/api/v1"
```

```python
# Creates the target URL, indicators, in this instance
target = base_url + "/indicators/"

# Get the response, which includes the first page of data as well as information on pagination and number of records
response = requests.get(target)

# Converts call into JSON
j = response.json()

# Converts JSON into a pandas DataFrame.
df_indicators = pd.json_normalize(j['data']) # pd.json_normalize flattens the JSON to accomodate nested lists within the JSON structure

# Loop until there are new pages with data
while j['nextPage'] != None:
    # Reset the target to the next page
    target = j['nextPage']

    #call the API for the next page
    response = requests.get(target)

    # Convert response to JSON format
    j = response.json()

    # Store the next page in a data frame
    df_temp = pd.json_normalize(j['data'])

    # Append next page to the data frame
    df_indicators = df_indicators.append(df_temp)
```

Read all pages

```python

params = {
    'page': 0,
    'indicator': 'stock_abs_+refug_host'}

# List to store data from all pages
all_data = []

while True:
    # Make API request for the current page
    response = requests.get(base_url, headers=header_, params=params)
    response_json = response.json()

    print(response_json["pagination"]["current_page"])
    print(response_json["pagination"]["total_pages"])

    # Append data from the current page
    observations = response_json.get("observations", [])
    all_data.extend(observations)

   # Check if there are more pages
    if params["page"] >= response_json["pagination"]["total_pages"]-1:
        break

    # Move to the next page
    params["page"] += 1

# Convert the aggregated data into a DataFrame
df_oim_complete = pd.DataFrame(all_data)

```
